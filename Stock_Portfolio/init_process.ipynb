# Import Module

```python
from datetime import datetime, timedelta, date
from pykrx import stock
import pandas as pd
import numpy as np

import pymysql
from sqlalchemy import create_engine 
```

# stock_price 데이터셋

## date_list에 2015년부터 날짜를 추가

```python
# 저장 시작 날짜 (2015년 1월 1일)

start_date = date(2015, 1, 1)

start_date
```

```python
# 현재 날짜
end_date = datetime.now()

end_date = end_date.date()

end_date
```

```python
# 날짜 계산
tdelta = end_date - start_date

tdelta.days
```

```python
# yyyymmdd 형식으로 1년간의 날짜 리스트 생성
date_list = [(start_date + timedelta(days=i)) for i in range(tdelta.days)]

date_list.insert(0, 'Date')
```

```python
date_list[-1]
```

## 현재 시장에 등록된 KOSPI 종목코드 반환

```python
tickers = stock.get_market_ticker_list(end_date)

len(tickers)
```

## 종목 코드에 해당하는 종목이름 반환

```python
companies = []

for ticker in tickers:
        company_name = stock.get_market_ticker_name(ticker)
        companies.append(company_name)

companies
```

```python
len(companies)
```

## 데이터프레임 만들기

```python
stock_price = pd.DataFrame(index = date_list[1:])
stock_price.index.name = date_list[0]

stock_price
```

## 각 종목당 2015년부터 지금까지의 주가 넣기

```python
for i in range(len(tickers)) :
    df = stock.get_market_ohlcv(date_list[1].strftime('%Y%m%d'), date_list[-1].strftime('%Y%m%d'), tickers[i])

    df_close = df['종가']
    df_close_dict = df_close.to_dict()

    stock_price[f'{companies[i]}'] = stock_price.index.map(df_close_dict).fillna(0)


stock_price
```

## 휴장일 제외

```python
# 모든 값이 0인 행 삭제
stock_price = stock_price[(stock_price != 0).any(axis=1)]

# 결과 출력
stock_price
```

## 인덱스를 DatetimeIndex로 변환

```python
stock_price.index = pd.to_datetime(stock_price.index)
```

## CSV 파일로 저장

```python
stock_price.to_csv('stock_price.csv', index = True, encoding = 'utf-8')
```

# stock_data 데이터셋

## 현재 시장에 등록된 KOSPI 종목코드 반환

```python
tickers = stock.get_market_ticker_list(end_date)

len(tickers)
```

## 종목 코드에 해당하는 종목이름 반환

```python
companies = []

for ticker in tickers:
        company_name = stock.get_market_ticker_name(ticker)
        companies.append(company_name)

len(companies)
```

## 1개월 전 대비 퍼센트 변화를 계산

```python
stock_price_2 = stock_price.resample('M').last().pct_change(1)

stock_price_2
```

## 연간 기대수익률, 연간 리스크 계산

```python
avg_returns = []
var_returns = []

for stock in stock_price_2.columns:
    stock_data = stock_price_2[stock]
    stock_data = stock_data.dropna()  # 결측값 제거
    stock_data = stock_data[(stock_data != np.inf)]
    stock_data = stock_data[(stock_data != -np.inf)]
    avg_returns.append(stock_data.mean() * 12)  # 월간 평균 수익률 * 12 (연간 기대수익률)
    var_returns.append(stock_data.var() * 12)  # 월간 분산 * 12 (연간 분산)
```

```python
stock_data = pd.DataFrame()

stock_data['코드'] = tickers
stock_data['기업명'] = companies
stock_data['예상수익률'] = avg_returns
stock_data['리스크'] = var_returns

stock_data
```

## CSV 파일로 저장

```python
stock_data.to_csv('stock_data.csv', index = True, encoding = 'utf-8')
```
